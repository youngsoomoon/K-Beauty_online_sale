```python
df_raw = pd.read_csv("data/국가_대륙_별_상품군별_온라인쇼핑_해외직접판매액_20210402140853.csv", encoding='cp949')
df_raw
```

```python
df = df_raw.melt(id_vars=["국가(대륙)별", "상품군별", "판매유형별"],
                 var_name="기간", value_name="백만원")
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df["기간"].map?
```

```python
df["연도"] = df["기간"].map(lambda x : int(x.split()[0]))
df.head()
```

```python
df["분기"] = df["기간"].map(lambda x : int(x.split()[1].split("/")[0]))
df.head()
```

```python
import numpy as np
```

```python
df["백만원"] = df["백만원"].replace("-",np.nan).astype(float)
df["백만원"]
```

```python
df.head()
```

```python
df = df[(df["국가(대륙)별"] != "합계") & 
        (df["상품군별"] != "합계") &
        (df["판매유형별"] != "계")].copy()
```

```python
sns.set(font_scale=1.0, font="Malgun Gothic")
df_fashion = df[df["상품군별"].str.contains("의류")]
plt.figure(figsize=(15, 4))
plt.title("의류 및 패션관련 상품")
plt.xticks(rotation=30)
sns.lineplot(data=df_fashion, x="기간", y="백만원", hue="국가(대륙)별")
```

```python
pivot = df_fashion.pivot_table(
    index="국가(대륙)별", values="백만원", 
    columns="연도", aggfunc="sum")
pivot
```

```python
plt.figure(figsize=(15, 7))
sns.heatmap(pivot, cmap="Blues", annot=True, fmt=".0f")
```
